```python
from rdkit.Chem import Descriptors
from rdkit.Chem import Lipinski

rdkit_funcs = {
    "exact_molecular_weight": Descriptors.ExactMolWt,
    "molecular_weight": Descriptors.MolWt,
    "logP":  Descriptors.MolLogP,
    "hydrogen_bond_donors": Lipinski.NumHDonors,
    "hydrogen_bond_acceptors": Lipinski.NumHAcceptors,
    "rotatable_bonds": Lipinski.NumRotatableBonds,
    "aromatic_rings": Lipinski.NumAromaticRings,
    "molar_refractivity": Descriptors.MolMR,
    "tpsa": Descriptors.TPSA
}
```

```python
from rdkit import Chem


data = []

for i, row in df.iterrows():
    m = Chem.MolFromSmiles(row["SMILES"])
    data.append({
         k:v(m) for k,v in rdkit_funcs.items()
                })

dfd = pd.DataFrame(data)
```

```python
dfd["molecule_name"] = df.reset_index()["Molecule name"]
```

```python
pd.merge(df, dfd, 'inn')
```
